###チャレンジ②当たり判定を変えてみる→当選確率を３％にしてみる

```python
import random
for i in range(100):
 a=random.randint(1,100)
 print(a)
 if(a==1 or a==2 or a==3):
  #a<4でも通る
  print('あたり')
 else:
  print('はずれ')
```

###チャレンジ③キャリーオーバーしたらどうなる→100連ガチャして当たらなかったら当選確率を上げる

```python
import random
pers=100
for i in range(100):
 a=random.randint(1,pers)
 print(a)
 if(a==1):
  #a<=persでもok
  print('あたり')
 else:
  pers=pers-1
  #a<=persの場合[pers=pers+1]となる
  #一増やす＝分子が増える＝確率が上がる
  print('キャリーオーバー^-^')
```
